My dataset is all police incidents that have occured since 2015.

```python
import pandas as pd
incidents = pd.read_csv("police.csv").sort_values("INCIDENTNEIGHBORHOOD").dropna()
List = incidents["INCIDENTNEIGHBORHOOD"].unique().tolist()

count = []
for r in range(98):
    count.append(0)
for r in range(len(incidents)):
    hoodName = incidents.iloc[r,9]
    i = List.index(hoodName)
    count[i]+=1
num = [[]]
for i in range(98):
    num[i].append(List[i])
    num[i].append(count[i])
    num.insert(i+1, [])

del num[63]
del num[63]
del num[63]
del num[88]
del num[94]

pd.set_option("display.max_rows", None, "display.max_columns", None)
table = pd.DataFrame(num, columns=["Neighborhood", "Police Incidents since 2015"])
table.sort_values("Police Incidents since 2015", ascending = True)
```
